# Write a Python program that prints the integers from 1 to 100.But for multiples of seven print "Fizz" instead of the number and for multiples of three     print "Buzz". For numbers that are multiples of three and seven, print "FizzBuzz".

```python
for i in range(101):
    if i % 3 == 0 and i % 7 == 0:
        print("FizzBuzz")
    elif i % 3 == 0:
        print("Buzz")
    elif i % 7 == 0:
        print("Fizz")
    else:
        print(i)
```
